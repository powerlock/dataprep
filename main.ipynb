# Smoothing method
* Savitzkyâ€“Golay filter: It is a process of convolution: fitting successive sub-sets of adjacent data points with a low degree polynominal by the method of linear least squares.

```python
from scipy.signal import savgol_filter
```

```python
x = np.linspace(0, 2*np.pi, 100)
y = np.sin(x) + np.cos(x) + np.random.random(100)
```

```python
plt.plot(x,y)
```

```python
for poly in range(3,6):
    for size in range(10,60,5):
        if poly < size:
            y_filtered = savgol_filter(y,size,poly)
            plt.plot(x,y_filtered, label='window-'+str(size)+'-'+str(poly)+' order')
            plt.legend(loc='lower left')
        else:
            pass
```

```python
y_filtered_op= savgol_filter(y,40,3)
y_model = np.sin(x) + np.cos(x) + 0
plt.plot(x,y,label='noise data')
plt.plot(x,y_filtered_op,label='filtered data')
plt.plot(x,y_model, label='model')
plt.legend()
```

This filter method helps to remove noise, so the curve is close to the original signal. The green curve above is the model with random constant value zero, therefore, it is lower than the original curve. The smoothed curve follows the noisy curve well. It is selected with different window size and exponential value.
